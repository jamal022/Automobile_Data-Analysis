```python
# As you can see there is no need for column index here so we gonna drop it.
df.drop('index',axis = 1 ,inplace = True)
```

```python
df['num-of-cylinders'].unique()
```

```python
# We can replace those string values by a new ones in int type.

new_num = {'four':4 , 'six':6 , 'five':5 , 'three':3 , 'twelve':12 , 'two':2 , 'eight':8}

df.replace({'num-of-cylinders':new_num},inplace=True)
```

```python
df.dtypes
```

```python
# Check Null Values 
df.isna().sum()
```

```python
# We can fill those null values with mean instead of drop them
df.fillna(df.mean(),inplace=True)
```

To display all company names with average mileage, we can group the data based on the 'company' and then selecting the column 'average-mileage' with mean() attribut to get the average mileage on the selected column

```python
company_ave_mileage['average-mileage'].mean()
```
